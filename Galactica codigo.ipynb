```python
import numpy as np
import matplotlib.pyplot as plt

G = 6.67430e-11 
M = 1.0 
a = 1.0 

def potencial_plummer(r, G, M, a):
    return -G * M / np.sqrt(r**2 + a**2)

def densidad_plummer(r, M, a):
    return (3 * M) / (4 * np.pi * a**3) * (1 + r**2 / a**2)**(-5/2)

r = np.linspace(0, 5, 250)

# Cálculo de las funciones
phi = potencial_plummer(r, G, M, a)
rho = densidad_plummer(r, M, a)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(r, phi, label='Potencial gravitatorio de Plummer')
plt.xlabel('r')
plt.ylabel('$\Phi(r)$')
plt.title('Potencial gravitatorio de Plummer')
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(r, rho, label='Densidad radial de Plummer', color='orange')
plt.xlabel('r')
plt.ylabel('$\\rho(r)$')
plt.title('Densidad radial de Plummer')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Definimos los arámetros
M = 1 
a = 1  

#Definimos las funciones para el potencial gravitatorio y la densidad de masa encontrada anteriormente
def potencial_gravitatorio(r, M, b):
    return - (M / a) * (1 / np.sqrt(1 + (r / a)**2))

def densidad_de_masa(r, M, b):
    return (3 * M) / (4 * np.pi * a**3) * (1 + (r / a)**2)**(-5/2)

r = np.linspace(0, 10, 400)

potencial = potencial_gravitatorio(r, M, b)
densidad = densidad_de_masa(r, M, b)

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('r')
ax1.set_ylabel('Potencial Gravitatorio', color=color)
ax1.plot(r, potencial, color=color, label='Potencial Gravitatorio')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper right')

ax2 = ax1.twinx()
color = 'tab:green'
ax2.set_ylabel('Densidad de Masa', color=color)
ax2.plot(r, densidad, color=color, label='Densidad de Masa')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper left')

fig.tight_layout()
plt.title('Potencial Gravitatorio y Densidad de Masa Radial de una Esfera de Plummer')
plt.show()
```
